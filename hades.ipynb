```python
import requests
import random
import time


def fetch_object(input):
    options = "https://rps101.pythonanywhere.com/api/v1/objects/all"
    response = requests.get(options)

    if response.status_code == 200:
        data = response.json()

        try:
            num = int(input)
            if 1 <= num <= len(data):
                return data[num - 1]
            else:
                raise IndexError("Invalid input. Number is out of range.")
        except ValueError:
            matching_objects = [obj for obj in data if obj.lower() == input.lower()]
            if matching_objects:
                return matching_objects[0]
            else:
                raise ValueError("Invalid input. Please enter a valid object name.")


def play_game(num):
    try:
        object_one = fetch_object(num)
        object_two = fetch_object(random.randint(1, 101))
        print(f"You have selected {object_one} computer has selected {object_two}")

        url = f"https://rps101.pythonanywhere.com/api/v1/match?object_one={object_one}&object_two={object_two}"
        response = requests.get(url)

        if response.status_code == 200:
            result = response.json()
            message = f"{result['winner'].capitalize()} {result['outcome']} {result['loser'].capitalize()}. {result['winner'].capitalize()} wins!"
            return message
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
            return None
    except Exception as e:
        return f"{e}"


def options():
    options_url = "https://rps101.pythonanywhere.com/api/v1/objects/all"
    options_response = requests.get(options_url)

    if options_response.status_code == 200:
        options = options_response.json()

        columns = 6
        rows = (len(options) + columns) // columns

        print("Options are:")
        for row in range(rows):
            for col in range(columns):
                index = row + col * rows
                if index < len(options):
                    print(f"{index + 1}. {options[index]:<15}", end="\t")
            print()
    else:
        print(f"Failed to fetch options. Status code: {options_response.status_code}")




play = True
print("Welcome to the game! Enter in a value in the field above")
print(options())
while play:
    val = input("Enter an object name or a number between 1 and 101 (enter q to quit): ")
    if val.lower() == 'q':
        play = False
    else:
        print(play_game(val))
        time.sleep(1)
```
